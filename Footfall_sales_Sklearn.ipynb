# Write a machine learing algorithm to predict daily salses based on footfalls in General store


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('General_Stores.csv')
print("Dataset preview")
data.head()
```

```python
X=data[['Footfalls']]
Y=data['DailySales']
```

```python
type(X)
```

```python
type(Y)
```

```python
plt.xlabel('Footfalls')
plt.ylabel('Daily Sales')
plt.scatter(X,Y,color='red',marker='o')
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
print(f"Coefficient (slope): {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
model.intercept_ + model.coef_*20
```

```python
model.intercept_ + model.coef_*35
```

```python
new_footfalls = pd.DataFrame({'Footfalls': [45, 50, 60]})
new_footfalls
```

```python
model.predict(new_footfalls)
```

```python


print("Predicted Daily Sales for new Footfalls:\n\n")

print(pd.DataFrame({'Footfalls': new_footfalls['Footfalls'], 'PredictedSales': model.predict(new_footfalls)}))
```
